# Brownian Motion Simulation in Python

This notebook simulates a single path of **standard Brownian motion**  using Python and visualizes the result.


## What is Brownian Motion?

Brownian motion $W(t)$ is a continuous-time stochastic process that satisfies:

1. $W(0) = 0$
2. $W(t) - W(s) \sim \mathcal{N}(0, t - s)$ for $t > s$
3. The increments are independent
4. The paths of $W(t)$ are continuous but nowhere differentiable

### Discrete Approximation

To simulate Brownian motion, we discretize time into $N$ steps over $[0, T]$, and generate the path using:

$$
W_{i+1} = W_i + \Delta W_i,\quad \Delta W_i \sim \mathcal{N}(0, \Delta t)
$$


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set parameters
T = 1.0             # Total time
N = 1000            # Number of time steps
dt = T / N          # Time step size
t = np.linspace(0, T, N+1)  # Time grid
plt.figure(figsize=(10, 4)) # Define the size of the graph
# Suppose we need 10 bm simulations
for i in range(10):
    # Generate normal increments ~ N(0, dt)
    dW = np.random.normal(0, np.sqrt(dt), size=N)

    # Construct Brownian path W(t)
    W = np.zeros(N+1)
    for i in range(N):
        W[i+1] = W[i] + dW[i]
    plt.plot(t, W, label='Simulated Brownian Motion Path')

# Modify the graph
plt.title('Single Brownian Motion Path')
plt.xlabel('Time $t$')
plt.ylabel('W(t)')
plt.grid(True)
plt.show()
```

## Summary

- We discretized the time interval and simulated normally distributed increments.
- The Brownian motion path is built by cumulatively summing the increments.
- This technique is the foundation for many stochastic models in quantitative finance, such as the **Geometric Brownian Motion (GBM)** used in the Black-Scholes model.
